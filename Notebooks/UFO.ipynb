#McNemar Chi-Square

```python
import pandas as pd
import statsmodels as sm
from statsmodels.stats.contingency_tables import mcnemar
from scipy import stats
```

```python
#Does the number of shapes change from 2013 to 2014?
```

```python
dataset2.info()
```

```python
sightings = pd.read_excel("../Data/Sightings.xlsx")
```

```python
sightings.head()
```

```python
import pandas as pd
import statsmodels as sm
from statsmodels.stats.contingency_tables import mcnemar
```

```python
sightings.info()
```

```python
def eval (series): 
    if series < 2000: 
        return 0
    if series >= 2000: 
        return 1
sightings['After_2000'] = sightings["Year"].apply(eval)
```

```python
sightings
```

```python
def avg (series): 
    if series >= 1082: 
        return 1
    if series <= 1082: 
        return 0
sightings['Greater_than_total_avg'] = sightings["Avg_seconds"].apply(avg)
```

```python
sightings
```

```python
sightings_crosstab = pd.crosstab(sightings['After_2000'], sightings['Greater_than_total_avg'])
sightings_crosstab
```

```python
result = sm.stats.contingency_tables.mcnemar(sightings_crosstab, exact=False, correction=True)
```

```python
print(result)
```

#The P value is greater which means the findings are not significant
